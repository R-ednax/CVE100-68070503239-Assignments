```python
#write your code
# --- Part 1: Basic Algebra and Substitution ---

# 1. Define symbolic variable and expression
import sympy as sp
x = sp.Symbol('x')
E_x = 3 * x**2+ 5 * x -7

print(f"E(x) ={E_x}")
```

```python
# 2. Evaluate (Substitute)
Value = 4

New_expression = E_x.subs(x, Value)

print(f"E(4) = {New_expression}")
```

Part 2: Solving Equations

Tasks:

Solve a Quadratic Equation: Use the solve() function to find the exact $x$ values that satisfy:

$$\\ x^2 - 4x + 3 = 0$$

$$$$

Solve a Linear System: Define two variables, $x$ and $y$. Use solve() to find the exact values of $x$ and $y$ that satisfy both equations:

$$\\ \text{Equation 1: } x + y = 10$$

$$$$$$
$$\text{Equation 2: } 2x - y = 5

$$\\ *(Hint: Pass a list of equations and a list of variables to `solve()`.)*$$

```python
#write your code
# --- Part 2: Solving Equations ---
# 1. Solve a Quadratic Equation: x^2 - 4x + 3 = 0
x = sp.Symbol("x")

Quad_Eq = x**2 - 4*x + 3 

Solutions_Qu = sp.solve(Quad_Eq, x)

print(f"Quadratic Equation = {Quad_Eq} = 0")
print(f"Solution for x = {Solutions_Qu}")
```

```python
# 2. Solve a Linear System
x, y = sp.symbols("x y")

eq1 = x + y - 10

linear_solutions = sp.solve([eq1], [x, y])

print(f"Equation solved: x + y = 10")
print(f"Solutions {x, y}: {linear_solutions}")


```

```python
#write your code
# --- Part 3: Fundamental Calculus ---
# 1. Solve function g(x)
x = sp.Symbol("x")

g_x = 4*x**3 - 6*x**2

Solutions_Qu = sp.solve(g_x, x)

print(f"Quadratic Equation = {Quad_Eq} = 0")
print(f"Solution for x = {Solutions_Qu}")
```

```python
# 1. Find the Derivative

x = sp.Symbol("x")

g_x = 4*x**3 - 6*x**2

g_prime_x = sp.diff(g_x, x)

print(f"Original Equation= {g_x}")
print(f"First Derivative= {g_prime_x}")
```

```python
# 2. Find the Indefinite Integral

integral_g_x = sp.integrate(g_x, x)

print(f"Indefinite integral of g(x): {integral_g_x} + C")
```
