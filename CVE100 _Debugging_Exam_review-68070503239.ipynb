```python
class Rectangle:
    def __init__(self,w,h):
        self.w = w
        self.h = h

    def area(self):
        area = self.w * self.h
        return area

rect = Rectangle(5, 3)
print("Area:", rect.area()) # Expected Output: Area: 15
```

Question 2: 10 points

```python
def even_numbers(nums):
    even_numbers = []
    for n in nums:
        if n % 2 == 0:
            even_numbers.append(n)
    return even_numbers

print(even_numbers([1,2,3,4,5,6])) # Expected Output: [2, 4, 6]
```

```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

a = Account("Alice", balance=100)
a.deposit(50)
print(a.balance) # Expected Output: 150
```

Question 4: 10 points

```python
class Student:
    def __init__(self, name, scores, score):
        self.name = name
        self.scores = scores
        self.score = score

    def add_score(self, score):
        self.scores.add(score)

    def avg(self):
        avg_score = sum(self.scores) / len(self.scores)
        if avg_score >= 80:
          grade = 'A'
        elif avg_score < 80:
          grade = 'B'
        else:
          grade = 'C'
        return avg_score , grade


students = [
    Student("A", [80, 70, 90], 85),
    Student("B", [60, 50, 65], 55)
]

for s in students:
    avg_score, grade = s.avg()
    print(s.name, "average:", avg_score)
    print(s.name, "grade:", grade)

# Expected Output:
# A average: 80.0
# A grade: A
# B average: 58.33
# B grade: B
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary #private attribute, do not change this.

    def bonus(self, percent):
        self.__salary = self.__salary + (self.__salary * percent/100)

    def get_salary(self):
        return self.__salary

e1 = Employee("John", 20000)
e2 = Employee("Jane", 25000)
e1.bonus(10)
e2.bonus(10)
print(e1.get_salary(), e2.get_salary()) # Expected Output: 22000 27500
```

```python
def write_data(filename):
    f = open(filename, "w")
    f.write("10 20 30 40")
    f.close()

def read_and_process(filename):
    with open(filename, "r") as f:
        data = f.read().split()
    total = 0
    #print(data)
    for x in data:
        total += float(x)
    avg = total / len(data)
    return avg

write_data("nums.txt") #file should be appeared on your computer
print("Average:", read_and_process("nums.txt"))
# Expected Output: Average: 25.0
```
